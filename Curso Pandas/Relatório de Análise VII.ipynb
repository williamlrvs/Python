## Criando Agrupamentos

```python
import pandas as pd
```

```python
dados = pd.read_csv('dados/aluguel_residencial.csv',sep=';')
dados.head(10)
```

```python
dados['Valor'].mean()
dados['Bairro']
```

```python
bairros = list(dados['Bairro'])
```

```python
selecao = dados['Bairro'].isin(bairros)
dados = dados[selecao]
```

```python
dados['Bairro'].drop_duplicates()
```

```python
grupo_bairro = dados.groupby('Bairro')
```

```python
type(grupo_bairro)
```

```python
grupo_bairro.groups
```

```python
for bairro, data in grupo_bairro:
    print('{} -> {}'.format(bairro,data.Valor.mean()))
```

```python
grupo_bairro[['Valor','Condominio']].mean().round(2)
```

## Estatísticas Descritivas

```python
grupo_bairro['Valor'].describe().round(2)
```

```python
grupo_bairro['Valor'].aggregate(['min','max','sum']).rename(columns = {'min': 'Minimo','max': 'Máximo', 'sum': 'Soma'})
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.rc('figure', figsize = (20,10))
```

```python
fig = grupo_bairro['Valor'].mean().plot.bar(color = 'blue')
fig.set_ylabel('Valor do Aluguel')
fig.set_title('Valor Médio do Aluguel por Bairro',{'fontsize': 22})
```

```python
fig = grupo_bairro['Valor'].max().plot.bar(color = 'blue')
fig.set_ylabel('Valor Maximo do Aluguel')
fig.set_title('Valor Médio do Aluguel por Bairro',{'fontsize': 22})
```

```python
fig = grupo_bairro['Valor'].std().plot.bar(color = 'blue')
fig.set_ylabel('Valor do desvio Padrão do Aluguel')
fig.set_title('Valor Médio do Aluguel por Bairro',{'fontsize': 22})
```

```python
fig = grupo_bairro['Valor'].min().plot.bar(color = 'blue')
fig.set_ylabel('Valor Minimo do Aluguel')
fig.set_title('Valor Médio do Aluguel por Bairro',{'fontsize': 22})
```
